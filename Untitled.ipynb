```python
import re

dic_bugs={
    "Lil Nas &":"Lil Nas X",
    "& Ambassadors":"X Ambassadors",
    "Mia &":"Mia X",
    "for KING & COUNTRY":"for KING and COUNTRY",
    "Brooks & Dunn":"Brooks and Dunn"
    
}

feat="Featuring"
def parse(nomes):
    
    new_nomes=nomes.replace("(","")
    new_nomes=new_nomes.replace(")","")
    new_nomes=new_nomes.replace("[","")
    new_nomes=new_nomes.replace("]","")
    new_nomes=new_nomes.replace('"',"")
    new_nomes=new_nomes.replace(", "," & ")
    new_nomes=new_nomes.replace("Featuring","&")
    new_nomes=new_nomes.replace("Feat.","&")
    new_nomes=new_nomes.replace(" Or "," & ")
    new_nomes=new_nomes.replace(" X "," & ")
    new_nomes=new_nomes.replace(" x "," & ")
    for i in dic_bugs:
        if i in new_nomes:
            new_nomes=new_nomes.replace(i,dic_bugs[i])
    return re.split(r'\s[,&+]\s', new_nomes)
    
    
print(parse('"Little" Jimmy Dickens'))
```

```python
def createmultifeat(lista):
    final=[]
    for i in range(len(lista)):
        for j in range(i+1,len(lista)):
            final.append([lista[i],lista[j]])
    return final
print (createmultifeat(['NAV', 'Meek Mill',"oi"]))
       
```

```python
def insertfeat(dic,lista):
    for i in range(len(lista)):
        change=0
        for j in dic:
            if (dic[j]["first"]==lista[i][0] and dic[j]["second"]==lista[i][1]) or (dic[j]["second"]==lista[i][0] and dic[j]["first"]==lista[i][1]):
                dic[j]["weight"]+=1
                change=1
        if (change==0):
            dic[len(dic)]={"first":lista[i][0],"second":lista[i][1],"weight":1}
        

test={0:{"first":'NAV',"second":"Meek","weight":1}}
l=[['Meek Mill','NAV']]
insertfeat(test, l)
print(test)
        
```
